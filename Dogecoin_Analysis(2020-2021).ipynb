# Return of Investment

**Q. How much profit/loss does a person make for each Dogecoin if he/she buys and sells their Dogecoin on an arbitrary day in the year of 2020?**

```python
date_bought=input("Enter the date when you bought dogecoins: ")
date_sold=input("Enter the date when you sold: ")
cp=(float)(df.loc[df['Date']==date_bought]['Average'])
sp=(float)(df.loc[df['Date']==date_sold]['Average'])
if(sp>cp):
  roi=(sp-cp)/cp
  print("Profit gained will be",roi*100,"% for one Dogecoin")
else:
  roi=(cp-sp)/cp
  print("Loss suffered will be",roi*100,"% for one Dogecoin")
```

# Prediction

**Q. How do you predict the fluctutation of Dogecoin from after atleast 30 days of the given dataset with the help of the given data**

```python
from fbprophet import Prophet
from fbprophet.plot import add_changepoints_to_plot, plot_cross_validation_metric
```

```python
prophet_df=df[['Date','Average']]
prophet_df.rename(columns={'Date':'ds','Average':'y'},inplace=True)
```


We created the Prophet instance with all default values, fit the dataset.

```python
prophet_basic = Prophet()
prophet_basic.fit(prophet_df[['ds','y']])
```

For predicting the values using Prophet, we created a dataframe with ds(datetime stamp) containing the dates for which we want to make the predictions. We used make_future_dataframe() to which we specify the number of days to extend into the future. By default it includes dates from the history.

```python
future= prophet_basic.make_future_dataframe(periods=30)
future.tail(2)
```

Plotting the predicted data

```python
forecast=prophet_basic.predict(future)
fig1 =prophet_basic.plot(forecast)
```

We also plotted the forecasted components to depict the trend and seasonality

```python
fig1 = prophet_basic.plot_components(forecast)
```

We added ChangePoints to Prophet. Changepoints are the datetime points where the time series have abrupt changes in the trajectory. 

```python
fig = prophet_basic.plot(forecast)
a = add_changepoints_to_plot(fig.gca(), prophet_basic, forecast)
```

Though this prediction may not be exactly accurate, we can still concur that the Dogecoin prices will continue to rise steadily.
